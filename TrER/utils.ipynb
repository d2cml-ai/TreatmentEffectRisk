```python
def summary_CVAR(data: pd.DataFrame, weights, IF, p, add_cols=None) -> pd.DataFrame:
    SW = data[weights] * data[IF]
    n, t = data.shape

    CVaR = np.nanmean(SW)
    std = np.nanstd(SW) / np.sqrt(n)

    result = {
        "p": [p],
        "CVAR": [CVaR],
        "Std": [std],
    }
    if add_cols is not None:
        for col in add_cols:
            result[col] = col
    data = pd.DataFrame(result)
    return result

def IF_CVaR(q, A_ipw_mu_2, tau_q, p):
    return q + (A_ipw_mu_2 -q) * tau_q / p
def IF_CVaR_puglin(q, tau,  p):
    return q + (tau - q) * (tau <=q) / p

def IF_CVaR_Bad( q, A_ipw_mu_2, tau_q, p):
    return q + (A_ipw_mu_2 -q) * ( tau_q <=q) / p

def IF_CVaR_ATE(q, A_ipw_mu_2, tau, p):
    tau_q = tau <= q
    return q + A_ipw_mu_2 * (tau_q / p - 1) - q * tau_q / p

def IF_CVaR_Bbound_mATE(A_ipw_mu_2, q, tau, b, p):
    muA_q = A_ipw_mu_2 - q
    muA_b = 0 ## todo: reemplazar el valor correct
    tau_b = tau - b
    return muA_q + muA_b * (tau_b < q) / 2/ p + (- muA_b * (-tau_b <= q)) / 2/p
    
```
